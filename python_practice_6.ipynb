# I. Docstrings.

1. Що таке docstrings? Навіщо вони потрібні?

Docstrings - це рядок документації, який розміщується в середині функції, модуля, класу або методу. Вони використовуються для того, щоб пояснити, як використовувати, які аргументи передавати, типи цих аргументів, повернені значення, можливі виключення, які можуть виникнути, приклади та інші деталі, які допомагають зрозуміти, як працює код.

4. Створіть docstring для функції, що перевіряє, чи є число простим. Увага: під час виклику функції ми очікуємо, що користувач(ка) може подати на вхід будь-який тип обʼєкту. У випадку, якщо на вході не підходящий тип даних, маємо кинути помилку.

```python
"""
This function checks if the number is prime.

Args:
    n (int): the number

Returns:
    bool: True if the number is prime, False otherwise

Raises:
    TypeError: if the input is not an integer
"""
```

5. Визначте функцію, яка перевіряє, чи є число простим, використовуючи документацію з завдання 4.

```python
def is_prime_number(n):
    """
    This function checks if the number is prime.
    
    Args:
        n (int): the number
    
    Returns:
        bool: True if the number is prime, False otherwise
    
    Raises:
        TypeError: if the input is not an integer
    """

    if not isinstance(n, int):
        raise TypeError("The input must be an integer")
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True
```

6. Створіть docstring для функції, що повертає найдовший спільний префікс двох рядків не залежно від регістру. Обовʼязково включіть два приклади згідно формату.

```python
"""
This function returns the longest common prefix of two strings regardless of the case.

Args:
    s1 (str): the first string
    s2 (str): the second string

Returns:
    str: the longest common prefix of two strings

Examples:
    >>> longest_common_prefix("flower", "flow")
    "flow"
    
    >>> longest_common_prefix("dog", "racecar")
    ""

Raises:
    TypeError: if the input is not a string
"""
```

7. Реалізуйте функцію згідно документації із завдання 6, використовуючи її.

```python
def longest_common_prefix(s1: str, s2: str) -> str:
    """
    This function returns the longest common prefix of two strings regardless of the case.
    
    Args:
        s1 (str): the first string
        s2 (str): the second string
    
    Returns:
        str: the longest common prefix of two strings
    
    Examples:
        >>> longest_common_prefix("flower", "flow")
        "flow"
        
        >>> longest_common_prefix("dog", "racecar")
        ""
    
    Raises:
        TypeError: if the input is not a string
    """

    if not isinstance(s1, str) or not isinstance(s2, str):
        raise TypeError("The input must be a string")
    i = 0
    while i < len(s1) and i < len(s2) and s1[i].lower() == s2[i].lower():
        i += 1
    return s1[:i]
```

8. Створіть docstring для функції, що розраховує суму щомісячного платежу по єОселі.

```python
"""
This function calculates the amount of the monthly payment for єОселя.

Args:
    amount (int | float): the amount of the loan
    initial_payment (int | float): the initial payment
    years (int | float): the number of years

Returns:
    float: the amount of the monthly payment

Examples:
    >>> calculate_monthly_payment(1_000_000, 280_000, 6)
    10000.0

Raises:
    ValueError: if the initial payment is greater than the amount of the loan or if the initial payment is less than 20% of the amount of the loan or if the number of years is greater than 20 or any of the arguments is negative
    TypeError: if the amount of the loan, the initial payment or the annual interest rate is not a number
"""
```

9. Використовуючи документацію із завдання 8, реалізуйте відповідну функцію.

```python
def calculate_monthly_payment(amount, initial_payment, years):
    """
    This function calculates the amount of the monthly payment for єОселя.

    Args:
        amount (int | float): the amount of the loan
        initial_payment (int | float): the initial payment
        years (int | float): the number of years

    Returns:
        float: the amount of the monthly payment

    Examples:
        >>> calculate_monthly_payment(1_000_000, 280_000, 6)
        10000.0

    Raises:
        ValueError: if the initial payment is greater than the amount of the loan or if the initial payment is less than 20% of the amount of the loan or if the number of years is greater than 20 or any of the arguments is negative
        TypeError: if the amount of the loan, the initial payment or the annual interest rate is not a number
    """

    if not all(map(lambda x: isinstance(x, (int, float)), [amount, initial_payment, years])):
        raise TypeError("The amount of the loan, the initial payment and the number of years must be numbers")

    if amount < 0 or initial_payment < 0 or years < 0:
        raise ValueError("The amount of the loan, the initial payment and the number of years must be positive")
    if initial_payment > amount:
        raise ValueError("The initial payment must be less than the amount of the loan")
    if initial_payment < 0.2 * amount:
        raise ValueError("The initial payment must be at least 20% of the amount of the loan")
    if years > 20:
        raise ValueError("The number of years must be less than 20")

    return (amount - initial_payment) / (years * 12)
```

10. Реалізуйте функцію, яка отримує значення зі словника за ключем та повертає його, а за умови відсутності ключа відловлює KeyError.

```python
def get_value_from_dict(d, key):
    """
    This function gets the value from the dictionary by key and returns it, and catches KeyError if the key is missing.
    
    Args:
        d (dict): the dictionary
        key: the key
    
    Returns:
        any: the value from the dictionary by key or None if the key is missing
    """
    
    try:
        return d[key]
    except KeyError:
        return None
```

11. Реалізуйте функцію із завдання 10 без відловлювання помилки, однак за допомогою конструкції if-else.

```python
def get_value_from_dict_2(d, key):
    """
    This function gets the value from the dictionary by key and returns it, and catches KeyError if the key is missing.
    
    Args:
        d (dict): the dictionary
        key: the key
    
    Returns:
        any: the value from the dictionary by key or None if the key is missing
    """
    
    if key in d:
        return d[key]
    else:
        return None
```

12. Реалізуйте функцію з завдання 10, використовуючи метод словника, який дозволяє зробити операцію навіть без if-else.

```python
def get_value_from_dict_3(d, key):
    """
    This function gets the value from the dictionary by key and returns it, and catches KeyError if the key is missing.
    
    Args:
        d (dict): the dictionary
        key: the key
    
    Returns:
        any: the value from the dictionary by key or None if the key is missing
    """
    
    return d.get(key)
```

13. Визначте функцію, яка обробляє значення, введене користувачем з терміналу, приводячи його до цілого числа. Використайте відловлення відповідної помилки. Наприклад, людина має ввести свій вік.

```python
def get_age():
    """
    This function processes the value entered by the user from the terminal, converting it to an integer.
    
    Returns:
        int: the age entered by the user
    """
    
    while True:
        try:
            age = int(input("Enter your age: "))
            return age
        except ValueError:
            print("The age must be an integer")
```

14. Визначте функцію з завдання 13, де використайте конструкцію if-else замість відловдення помилки.

```python
def get_age_2():
    """
    This function processes the value entered by the user from the terminal, converting it to an integer.
    
    Returns:
        int: the age entered by the user
    """
    
    while True:
        age = input("Enter your age: ")
        if age.isdigit():
            return int(age)
        else:
            print("The age must be an integer")
```

15. Визначте функцію з завдання 13, враховуючи додаткове обмеження, що число має бути у діапазоні від 18 до 120 (повнолітня людина). Відловіть відповідні типи помилок.

```python
def get_age_3():
    """
    This function processes the value entered by the user from the terminal, converting it to an integer.
    
    Returns:
        int: the age entered by the user
    """
    
    while True:
        try:
            age = int(input("Enter your age: "))
            if 18 <= age <= 120:
                return age
            else:
                print("The age must be in the range from 18 to 120")
        except ValueError:
            print("The age must be an integer")
```
